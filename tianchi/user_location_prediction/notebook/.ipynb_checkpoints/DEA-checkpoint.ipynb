```python
wifi = pd.read_csv('./preprocessing_data/sample_wifi.csv')
sample = pd.read_csv('./preprocessing_data/sample.csv')
shop = pd.read_csv('./preprocessing_data/shop.csv')
```

### single mall 

```python
f, ax = plt.subplots()
ax.scatter(shop[shop.mall_id=='m_690'].lgt-122,
           shop[shop.mall_id=='m_690'].ltt-31, marker='o',alpha=0.6,s=30,label='shop')
ax.set_xlim([0.345,0.349])
ax.set_ylim([0.833,0.836])
longitude_ic = (shop[shop.mall_id=='m_690'].lgt-122).mean()
latitude_ic = (shop[shop.mall_id=='m_690'].ltt-31).mean()
ax.scatter(longitude_ic, latitude_ic,c='r', marker='^',s=40, label='mall')
ax.legend()
```
