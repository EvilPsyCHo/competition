```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import statsmodels as sm
from sklearn.metrics import pairwise, mutual_info_score

%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./bank-additional-train.csv')
```

```python
MI = {col: mutual_info_score(data[col], data.y) for col in data.columns[:-1]}
```

```python
f, ax = plt.subplots(figsize=(16, 8))
pd.Series(MI).sort_values(ascending=False).plot(kind='bar', ax=ax, fontsize=18)
ax.set_title('Mutual Information score between "X" and "Y"', size=20)
```
